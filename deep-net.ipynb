```python
import os

#declare train and validation dirs
base_dir = '/kaggle/input/cats-and-dogs-filtered/cats_and_dogs_filtered'
train_dir = os.path.join(base_dir, 'train')
valid_dir = os.path.join(base_dir, 'validation')

train_cat_dir = os.path.join(train_dir, 'cats')
train_dog_dir = os.path.join(train_dir, 'dogs')

validation_cat_dir = os.path.join(valid_dir, 'cats')
validation_dog_dir = os.path.join(valid_dir, 'dogs')

print('done!')
```

```python
# importanto biblioteca tensorflow
import tensorflow as tf

# usamos a biblioteca keras que possui varios modelos prontos
# configuramos uma rede convolucional - CNN
model = tf.keras.models.Sequential([
    # conv2d é a primeira camada, devemos especificar o tamanho da entrada
    tf.keras.layers.Conv2D(16, (3, 3), activation='relu', input_shape=(150, 150, 3)),
    tf.keras.layers.MaxPooling2D(2, 2),
    # repete o processo
    tf.keras.layers.Conv2D(32, (3, 3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2, 2),
    tf.keras.layers.Conv2D(64, (3, 3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2, 2),
    # achata resultado para colocar nas camandas densas
    tf.keras.layers.Flatten(),
    # define 512 neuronios para processar o output das camadas anteriores
    tf.keras.layers.Dense(512, activation='relu'),
    # camada de output única. resultado é 0 se gato e 1 se cachorro
    tf.keras.layers.Dense(1, activation='sigmoid') # sigmoid funciona para problemas binarios, por isso usamos essa
])

print('done!')
```

```python
# show some model info
model.summary()

# modelo relativamente simples mas tem 9mi params
```

```python
'''from tensorflow.python import metrics # essa lib da pau e nao baixa, pois n existe
model.compile(optimizer='adam',
             loss='binary_crossentropy',
             metrics=['accuracy'])'''
model.compile(optimizer='adam',
             loss='binary_crossentropy')

print('done!')
```

```python
from tensorflow.python.ops.batch_ops import batch
from tensorflow.keras.preprocessing.image import ImageDataGenerator

#reescalando imagens com o parametro de reescalamento
train_datagen = ImageDataGenerator(rescale=1.0/255)
test_datagen = ImageDataGenerator(rescale=1.0/255)

# cria gerador com diretorio usando flow_from_directory
train_generator = train_datagen.flow_from_directory(train_dir, batch_size=20, class_mode='binary',
                                                   target_size=(150, 150))
validation_generator = test_datagen.flow_from_directory(valid_dir, batch_size=20, class_mode='binary',
                                                   target_size=(150, 150))
```

```python
history = model.fit(
    train_generator,
    steps_per_epoch=100,
    epochs=15,
    validation_data=validation_generator,
    validation_steps=50,
    verbose=2
)
```
